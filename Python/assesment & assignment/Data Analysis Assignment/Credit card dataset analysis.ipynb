# 10. Analyse spending pattern based on age groups and avg spending amount for each group

```python
from datetime import datetime
df2=df[['dob','amt']].copy()
df2['dob']=pd.to_datetime(df2['dob'])
current_year=datetime.now().year
df2["Age"]=current_year-df2['dob'].dt.year
df2
```

```python
bins=[18,25,50,100]
labels = ["Young (18-25)", "Middle-aged (26-50)", "Senior (51+)"]
```

```python
df2["age_groups"]=pd.cut(df2['Age'],bins=bins,labels=labels)
df2
```

```python
#Average spending amount for each group
df2.groupby("age_groups",observed=False)['amt'].mean()
```

```python
df2.groupby("age_groups",observed=False)["amt"].agg(["sum", "mean", "count"])
```
